##  Bonus Question

```python
a=int(input())
b=int(input())
x=input()
if x=="+":
    print(a+b)
elif x=="-":
    print(a-b)
elif x=="*":
    print(a*b)
elif x=="/":
    print(a/b)
elif x=="//":
    print(a//b)
elif x=="**":
    print(a**b)
else:
    print(a%b)
```
