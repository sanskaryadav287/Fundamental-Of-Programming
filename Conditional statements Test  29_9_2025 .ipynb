## 1 Check if day is weekend or weekday

```python
a="Sunday"
if a=="Monday" or a== "Tuesday" or a=="Wednesday" or a=="Thursday" or a=="Friday":
  print("Weekday")
else:
  print("Weekend")
```

## 2 Find profit or Loss

```python
MRP=int(input("Enter a MRP: "))
SP=int(input("Enter Selling amount: "))
if MRP<SP:
    print("Profit")
elif MRP==SP:
    print("No Profit and no loss")

else:
    print("Loss")
```

## 3 Check Multiple of 7
