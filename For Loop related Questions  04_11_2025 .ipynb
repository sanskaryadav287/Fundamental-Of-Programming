```python
##  1 Find The Largest Number in the List [4,9,2,10,7]  using for Loop ###
x=[4,9,2,10,7] 
largest=x[0]
for i in x:
    if i>largest:
        largest=i
print(largest)
```

```python
###   2 count how many even and odd numbers are present in [1,2,3,4,5,6].   ###
l1=[1,2,3,4,5,6]
even_count=0
odd_count=0
for i in l1:
    if i%2==0:
        even_count+=i
    else:
        odd_count+=i
print(even_count)
```

```python
##  3. use a for loop to create a new list containing squares of numbers from 1 to 5 ###
sqre=[]
for i in range(1,6):
    sqre.append(i**2)
print(sqre)
```
