```python
##   series Two list by For Loop  ##
a=["Red","Yellow","Green"]
b=["Apple","Banana","Grapes"]
for i in a:
    for u in b:
        print(i,u)
```

```python
a=["Red","Yellow","Green"]
b=["Apple","Banana","Grapes"]
for i in a:
    for u in b:
         if i=="Yellow":
                continue
         elif u=="Banana":
             continue
         print(i,u)
```
